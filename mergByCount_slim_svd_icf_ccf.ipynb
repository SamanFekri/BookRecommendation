```python
svd_db_tokenize = []
for i in range(7944):
    t = svd_db.iloc[i][1].split()
    svd_db_tokenize.append(t)
```

```python
icf_db_tokenize = []
for i in range(7944):
    t = icf_db.iloc[i][1].split()
    icf_db_tokenize.append(t)
```

```python
ccf_db_tokenize = []
for i in range(7944):
    t = ccf_db.iloc[i][1].split()
    ccf_db_tokenize.append(t)
```

```python

def hybrid(a,b,c,d):
    output = []

    for user_index in range(0 ,7944):
        temp = ''
        counter = 0
        for sell_index in range(0,10):
            sell = slim_db_tokenize[user_index][sell_index]
            if counter != a:
                temp = temp +' '+ str(sell)
                counter += 1

        counter = 0
        for sell_index in range(0,10):
            sell = svd_db_tokenize[user_index][sell_index]
            if counter != b:
                temp = temp +' '+ str(sell)
                counter += 1
        
        counter = 0
        for sell_index in range(0,10):
            sell = icf_db_tokenize[user_index][sell_index]
            if counter != (c):
                if sell not in temp:
                    temp = temp +' '+ str(sell)
                    counter = counter + 1
                    
        counter = 0
        for sell_index in range(0,10):
            sell = ccf_db_tokenize[user_index][sell_index]
            if counter != (d):
                if sell not in temp:
                    temp = temp +' '+ str(sell)
                    counter = counter + 1              
                    
        output.append(temp[1:])
    return output
    
```
